```python
import cv2
from darkflow.net.build import TFNet
import matplotlib.pyplot as plt
import numpy as np
```

```python

options = {"model": "cfg/yolo-voc-test-cigarette.cfg", 
           "batch": 2,
           "epoch": 2,
           "load" : -1,
           "gpu": 0.5,
           "lr" : 1e-6,
           "train": True,
           "annotation": "./data/annotations_cigarette/",
           "dataset": "./data/dataset"}


tfnet = TFNet(options)
```

```python
tfnet.train()
```

## 테스트용 options

```python
options = {
    'model' : "cfg/yolo-voc-test-cigarette.cfg",
    'load' : 38076,
    'threshold' : 0.1,
    'gpu' : 0.5
}

tfnet = TFNet(options)

```

## 영상 테스트

```python
cap = cv2.VideoCapture('test.mp4')

width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# 비디오 코덱은 XVID
fourcc = cv2.VideoWriter_fourcc(*'XVID')
# 30프레임짜리 XVID코덱으로 저장하는 포맷 변수 선언
writer = cv2.VideoWriter('output_cigga3.avi', fourcc, 30.0, (width, height))
while(cap.isOpened()) :
    ret, frame = cap.read()
    if ret == False:
        break
    result = tfnet.return_predict(frame)
    result_img = frame.copy()
    # 모자이크 처리 부분
    if result:
        print(result)
        for res in result:
            x = res['topleft']['x']
            y = res['topleft']['y']
            w = res['bottomright']['x']
            h = res['bottomright']['y']
            target = frame[y:h, x:w]
            flt = cv2.blur(target, (10, 10))  # blur처리 한 후 flt 변수에 저장
            result_img[y:h, x:w] = flt
            tl = (x, y)
            br = (w ,h)
            label = res['label']
            conf = res['confidence']
            result_img= cv2.rectangle(result_img, tl, br, (0, 255, 0), 2)
            result_img= cv2.putText(result_img, label, tl, cv2.FONT_HERSHEY_TRIPLEX, 1, (0,0,0), 2)
        
    else :
        result_img = frame.copy()
        

    writer.write(result_img)

    if cv2.waitKey(1)&0xFF == 27:
        break
cv2.destroyAllWindows()
```
